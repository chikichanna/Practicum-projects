## Задания

### Сколько книг вышло после 1 января 2000 года

```python
query = '''SELECT COUNT(*)
           FROM books
           WHERE publication_date > '2000-01-01'
        '''
print('Количество книг:',pd.io.sql.read_sql(sql=text(query), con = con).squeeze())
```

Всего в базе 1000 книг, так что большинство из них вышло в 21 веке

### Количество обзоров и средняя оценка каждой книги

```python
query = '''WITH 
           rev AS(SELECT book_id,
                  COUNT(review_id) AS reviews_count
                  FROM reviews
                  GROUP BY book_id),
           rate AS(SELECT book_id,
                   AVG(rating) AS avg_rating
                   FROM ratings
                   GROUP BY book_id)
           SELECT b.title,
                  rev.reviews_count,
                  rate.avg_rating
           FROM books AS b
           LEFT JOIN rev ON rev.book_id=b.book_id
           JOIN rate ON rate.book_id=b.book_id
           ORDER BY rev.reviews_count DESC, rate.avg_rating DESC
           '''
pd.io.sql.read_sql(sql=text(query), con = con)
```

Больше всего отзывов на Сумерки

### Издательство, которое выпустило наибольшее число книг

Наибольшее число книг толще 50 страниц чтобы исключить из анализа брошюры

```python
query = '''SELECT p.publisher,
                  COUNT(book_id)
           FROM books
           JOIN publishers AS p ON p.publisher_id=books.publisher_id
           WHERE num_pages > 50 
           GROUP BY p.publisher_id
           ORDER BY COUNT(book_id) DESC
           LIMIT 1
           '''
pd.io.sql.read_sql(sql=text(query), con = con)
```

Наибольшее число книг из нашей базы было выпущено издательством Penguin Books - 42 книги. Всего в нашей базе записано 340 издательств, но выпускали книги (не брошюры) из них 334.

### Автор с самой высокой средней оценкой книг

Учитываются только книги с 50 и более оценками

```python
query = '''WITH 
           rate AS(SELECT book_id,
                   AVG(rating) AS avg_rating
                   FROM ratings
                   GROUP BY book_id
                   HAVING count(book_id) >= 50)
           SELECT a.author,
                  AVG(rate.avg_rating)
           FROM books AS b
           JOIN rate ON rate.book_id=b.book_id
           JOIN authors AS a ON a.author_id=b.author_id
           GROUP BY a.author
           ORDER BY AVG(rate.avg_rating) DESC
           LIMIT 1
           '''
pd.io.sql.read_sql(sql=text(query), con = con)
```

Самым высокооценным автором стала Дж.К.Роулинг. В нашей базе 636 автора, но только 14 из них имеют книги с 50 и более оценок

### Cреднее количество обзоров от пользователей, которые поставили больше 48 оценок

```python
query = '''WITH 
           rev AS(SELECT username,
                  COUNT(review_id) AS reviews_count
                  FROM reviews
                  GROUP BY username),
           rate AS(SELECT username
                   FROM ratings
                   GROUP BY username
                   HAVING count(rating_id) > 48)
           SELECT AVG(rev.reviews_count)
           FROM rate
           JOIN rev ON rate.username=rev.username
           '''
print('Среднее число обзоров:',pd.io.sql.read_sql(sql=text(query), con = con).squeeze())
```

Рейтинги оставляло 160 пользователей, как и обзоры. Но только 13 из них поставили больше 48 оценок. 
Эти позователи в среднем оставляли по 24 обзора.
