<b>Описание данных:</b>
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Предобработка данных

#### Замена регистра в названиях столбцов

```python
df.columns = df.columns.str.lower() #переименуем столбцы в нижний регистр
```

```python
#check
df.columns
```

#### Замена типа данных

```python
df['year_of_release'] = df['year_of_release'].astype('Int64') #преобразуем в int поддерживающий пропущенные значения
df['critic_score'] = df['critic_score'].astype('Int64')
```

Поменяли типы данных в столбцах с данными о годе выпуска и об оценке критиков. В столбце с оценками пользователей присутствует значение строковое tbd, означающее "to be defined" (подлежит определению), что мешает преобразовать тип данных в вещественные числа.

```python
df.loc[df['user_score'] == "tbd",'user_score'] = np.nan #заменяем "tbd" на пропуск
df['user_score'] = df['user_score'].astype('float')
```

#### Обработка пропущенных данных.

```python
def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (15,5), legend = False, fontsize = 16, xlim=(0,100))
            .set_title('Процент пропусков \n', fontsize = 22) 
        );    
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции ')
```

```python
pass_value_barh(df)
```

```python
df = df.query('~name.isna()') #удаляем строки c пропуском в названии, у  тех же пропущен жанр. Таких строк всего две.
```

```python
df = df.query('~year_of_release.isna()') #удаляем строки с пропуском в годе выпуска. Таких строк немного, всего 269. 
                                         #Нет смысла их заполнять, это может исказить данные
```

Пропущенные значения в строках с оценками нет смысла заполнять сгенерированными данными, так как это может исказить картину исследования.

```python
df['rating'] = df['rating'].fillna('unknown') #если неизвестен рейтинг игры, то поставим unknown
```

```python
df.duplicated().sum()
```

Полных дубликатов в датасете нет.

#### Создание нового столбца с суммарными продажами во всех регионах

```python
df['sales_total'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

```python
#check
df.head(10)
```

### Исследовательский анализ данных

#### Сколько игр выпускалось в разные годы

```python
(
    (df['year_of_release'].value_counts(sort=False).sort_index(ascending=False))
    .to_frame()
    .plot(kind = 'barh', figsize = (15,10), legend = False, fontsize = 16, xlim=(0,1500),grid =True)
    .set_title('Год выпуска \n', fontsize = 22) 
);    
```

До 1994 выпускалось меньше 100 игр в год. C 2008 года количество выпускаемых игр сокращается

#### Как менялись продажи по платформам

```python
platform_years = df.pivot_table(index='platform', columns='year_of_release',values='sales_total', aggfunc='sum')
platform_years['total'] = platform_years.sum(axis=1)
platform_years = platform_years.sort_values('total').loc[:,platform_years.columns != 'total']
plt.figure(figsize=(20,10))
sns.heatmap(platform_years, annot =True, fmt=".0f", linewidths=.5)
plt.title('Продажи по годам и платформам, млн. копий \n', fontsize = 22);
```

На данной диаграмме видно, что средний срок жизни платформы около 10 лет. 
Самая долгоживущая платформа для игр это ПК, но с 2011 года видна тенденция на снижение ее популярности. 
Две самые молодые платформы это PS4 и Xbox One.
Все платформы отсортированы по суммарному количеству проданных копий за все время, поэтому наиболее интересными для исследования представляются платформы в нижней части диаграммы.
Заметна тенденция на увеличение количества активных платформ на рынке с течением времени.

```python
plt.figure(figsize=(15,5))
(
    platform_years.count().plot(title='Количество активных платформ \n',fontsize=20)
    .set_xlabel('год')
    
);
```

Так же на тепловой карте заметна аномалия, в виде игры на Nintendo DS, с годом выпуска в 1985 году, когда платформы еще не существовало.

```python
df = df.drop(df[df['name'] == 'Strongest Tokyo University Shogi DS'].index) #удалим эту строчку, чтобы она не мешала дальнейшему анализу
platform_years = df.pivot_table(index='platform',   #переопределим датасет для вывода графика
                                columns='year_of_release',values='sales_total', aggfunc='sum')
```

На графике видно, что к 2017 году полностью угасли платформы Nintendo DS, Playstation 2 и Portable Playstation.
Почти угасли Playstation 3, Playstation Vita, Nintendo Wii и Xbox 360.

- ПК, как было сказано ранее, самая долоиграющая платформа на рынке, что и понятно. Компьютеры более многофунциональны. Скорее всего выпуск новых игровых консолей будет продолжать уменьшать долю рынка игр на ПК, но с ростом самого рынка количество проданных копий может вырасти. 
- Nintendo Wii U угасает, но может быть еще пару лет на нее что-то будет выпускаться.

На графике заметно четкая сменяемость консолей одного производителя. Так у лидеров продаж Microsoft и Sony Xbox One сменил Xbox 360, а Playstation 3 сменил Playstation 4 соответственно.
У еще одного лидера продаж - Nintendo, в наших данных нет следующего поколения, но учитывая угасание 3DS (на 3 месте по продажам за 3 года) оно очень вероятно.  
За 2016 год данные выгружены не целиком, поэтому несмотря на кажущееся падение продаж в 2016 году у  Xbox One и PS4, они еще могут вырасти дальше, пока не появится следующее поколение.


```python
grouped = (relevant.loc[:,['platform', 'sales_total']]
           .groupby(['platform'])
           .median()
           .sort_values(by='sales_total')) #Сгруппируем и отсортируем по медиане общих продаж по платформе

plt.figure(figsize=(15,10))
sns.boxplot(data=relevant, x='sales_total', y='platform', order=grouped.index)
plt.xlim([0,2]);
plt.title('Продажи по платформам, млн. копий \n', fontsize = 22);
```

Этот график говорит нам, о том как на разных платформах продаются игры. График отсортирован от меньшему к большему и мы видим, что больше всего копий медианно продавалось за этот период для платформы Xbox One.  
Так же не далеко отстает Playstation 4. 
Между ними затесалась Nintendo Wii U. Несмотря на то, что на эту консоль было выпущено значительно меньше игр(73 против 376 у PS4), что повлияло на общее количество копий, они продаются ничем не хуже конкурентов.

#### Влияние отзывов на продажи на платформе Xbox One

```python
corr = relevant.query('platform == "XOne"')[[
    'na_sales','eu_sales','jp_sales','other_sales','critic_score', 'user_score','sales_total']].corr() #считаем коэффициент корреляции
plt.figure(figsize=(10,10))
sns.heatmap(corr,square=True, cmap="coolwarm",annot =True, fmt=".2f",vmin=-1, vmax=1, linewidth=.5) #строим тепловую карту
plt.title('Корреляция оценок с продажами \n', fontsize = 22);
```

```python
sns.pairplot(relevant.query('platform == "XOne"')[[
    'critic_score', 'user_score','sales_total']]).fig.suptitle("Корреляция оценок с продажами", y=1.01)
plt.gcf().set_size_inches(15,8)
```

На тепловой карте наблюдается небольшая положительная корреляция между продажами и оценками критиков, причем эта корелляция меньше по продажам в Японии. По оценкам пользователей же корреляция практически нулевая. 

#### Корреляция оценок критиков с продажами для других платформ

```python
for i in ('WiiU','PS4','X360','PS3','3DS','PC','PSV'):
    relevant[relevant['platform'] == i].plot.scatter(x='critic_score', y='sales_total', alpha=0.5, c='SteelBlue', title=i);
    print('Коэффициент корреляции для', i,
          np.round((relevant[relevant['platform'] == i].corr()).loc['critic_score','sales_total'],decimals = 2))
```

Как показывают графики и коэффициенты, корреляция между продажами и оценками критиков и на других платформах слабая (<0.7).

### Проверка гипотез

#### Гипотеза 1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевая гипотеза:  
    Средние двух генеральных совокупностей пользовательских рейтингов на платформах Xbox One и PC равны между собой

Альтернативная гипотеза:   
    Они не равны

```python
XOne_user_score = relevant.query('platform == "XOne" & 0 < user_score')['user_score'] \
                       .to_frame() # sample_1 - выборка с пользовательскими оценками игр на Xbox One
PC_user_score = relevant.query('platform == ("PC") & user_score >0')['user_score'] \
                        .to_frame() # sample_2 - выборка с пользовательскими оценками игр на ПК

alpha = .05 # уровень статистической значимости

results = st.ttest_ind(XOne_user_score, PC_user_score, equal_var = False)


print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Вероятность случайно получить такое или большее различие равно около 11% при условии, что нулевая гипотеза верна. Это слишком большая вероятность, чтобы делать вывод о значимом различии.

#### Гипотеза 2: Средние пользовательские рейтинги жанров Action  и Sports  разные

Нулевая гипотеза:  
    Средние двух генеральных совокупностей пользовательских рейтингов игр в жанре Экшен и игр в жанре Спорт равны между собой

Альтернативная гипотеза:   
    Они не равны

```python
action_user_score = relevant.query('genre == "Action" & 0 < user_score')['user_score'] \
                       .to_frame() # sample_1 - выборка с пользовательскими оценками игр в жанре Экшен
sports_user_score = relevant.query('genre == "Sports" & user_score >0')['user_score'] \
                        .to_frame() # sample_2 - выборка с пользовательскими оценками игр в жанре Спорт

alpha = .05 # уровень статистической значимости

results = st.ttest_ind(action_user_score, sports_user_score, equal_var = False)


print('p-значение: ', np.round(results.pvalue,decimals=15))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Вероятность случайно получить такое или большее различие равно сильно меньше 1% при условии, что нулевая гипотеза верна .Это явно слишком маленькая вероятность, делаем вывод о значимом различии генральных совокупностей.   
Гипотезе о разности совокупностей данные не противоречат.
